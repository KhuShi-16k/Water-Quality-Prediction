```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

# Load the data
df = pd.read_csv('data.csv')

# Convert Trihalomethanes to float
df['Trihalomethanes'] = pd.to_numeric(df['Trihalomethanes'], errors='coerce')

# Check for any null values after conversion
print(df.isnull().sum())

# Handle missing values
imputer = SimpleImputer(strategy='mean')
df[['Trihalomethanes']] = imputer.fit_transform(df[['Trihalomethanes']])

# Feature scaling
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df.drop(columns=['Potability']))

# Create a new DataFrame with scaled features
scaled_df = pd.DataFrame(scaled_features, columns=df.columns[:-1])

# Add the target variable back to the scaled DataFrame
scaled_df['Potability'] = df['Potability'].values

# Save the cleaned and scaled data to a new CSV file
scaled_df.to_csv('newdata.csv', index=False)

print(scaled_df.head())
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from imblearn.over_sampling import SMOTE
import numpy as np
import joblib

# Load the data
data = pd.read_csv('newdata.csv')

# Define features and target
X = data.drop('Potability', axis=1)
y = data['Potability']

# Removing outliers using IQR for each numerical column
for column in X.columns:
    Q1 = X[column].quantile(0.25)
    Q3 = X[column].quantile(0.75)
    IQR = Q3 - Q1
    outlier_condition = ~((X[column] < (Q1 - 1.5 * IQR)) | (X[column] > (Q3 + 1.5 * IQR)))
    X = X[outlier_condition]
    y = y[outlier_condition]

# Apply SMOTE for balancing classes
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# Initialize the RandomForestClassifier
model = RandomForestClassifier(random_state=42)

# Define an expanded parameter grid for tuning
param_distributions = {
    'n_estimators': np.arange(50, 301, 50),
    'max_depth': [None] + list(np.arange(10, 61, 10)),
    'min_samples_split': [2, 5, 10, 15],
    'min_samples_leaf': [1, 2, 4, 6],
    'max_features': ['sqrt', 'log2', None],
}

# Set up RandomizedSearchCV
random_search = RandomizedSearchCV(estimator=model, param_distributions=param_distributions,
                                   scoring='accuracy', cv=10, n_iter=100,
                                   verbose=2, random_state=42, n_jobs=-1)

# Train the model with RandomizedSearchCV
random_search.fit(X_train, y_train)

# Get the best model
best_model = random_search.best_estimator_

# Save the model
joblib.dump(best_model, 'water_potability_model.pkl')

# Predictions with the best model
y_pred = best_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

# Print evaluation metrics
print("Best Parameters:", random_search.best_params_)
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("Confusion Matrix:\n", conf_matrix)
```
